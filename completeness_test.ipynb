## Checking if the whole dataset is in the folder

```python
import pandas as pd
import os
from PIL import Image 
```

## Read the dataset csv

```python
folder_path = "data/scene1"
files = []
for i in os.listdir(folder_path):
    if os.path.isfile(os.path.join(folder_path,i)) and 'Dataset' in i:
        files.append(i)

dataset = pd.read_csv(os.path.join(folder_path, files[0]))
dataset.head()
```

```python
files_not_found = []
for image_name in dataset['rendered_image']:
    try:
        Image.open(os.path.join(folder_path, image_name))
    except IOError:
        files_not_found.append(image_name)

print(files_not_found)
```
